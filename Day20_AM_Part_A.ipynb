# 1. Max, Min, Frequency using loops

```python
data = [4, 2, 7, 2, 9, 4, 1, 7]

max_val = data[0]
min_val = data[0]

for x in data:
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x

print("Max:", max_val)
print("Min:", min_val)


freq = {}

for x in data:
    if x in freq:
        freq[x] += 1
    else:
        freq[x] = 1

print("Frequency:", freq)
```

# 2. While loop tasks

```python
# Reverse number
num = 123
rev = 0

temp = num
while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

print("Reversed:", rev)

# Palindrome check
num = 121
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if num == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# 3. Tuples and Dictionaries

```python
pairs = [("a", 10), ("b", 20), ("c", 15)]

# Convert to dictionary
d = {}

for pair in pairs:
    key = pair[0]
    value = pair[1]
    d[key] = value

print("Dictionary:", d)

# Find key with highest value
max_key = None
max_value = None

for key in d:
    if max_value is None or d[key] > max_value:
        max_value = d[key]
        max_key = key

print("Max key:", max_key)
```



# 4. *args function

```python
def sum_avg(*args):
    total = 0
    count = 0

    for x in args:
        total += x
        count += 1

    avg = total / count

    return total, avg

print(sum_avg(10, 20, 30, 40))
```

# 5. **kwargs function

```python
def top_student(**kwargs):
    best_name = None
    best_marks = None

    for name in kwargs:
        if best_marks is None or kwargs[name] > best_marks:
            best_marks = kwargs[name]
            best_name = name

    return best_name, best_marks

print(top_student(Alice=85, Bob=92, Charlie=88))
```
